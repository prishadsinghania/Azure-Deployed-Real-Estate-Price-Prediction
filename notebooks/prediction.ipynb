```python
import pandas as pd
import pickle
```

```python
# Load the unseen data
ds = pd.read_csv('RealEstatePrediction.csv')
ds.info()
```

```python
# Open the pickle file for reading in binary mode
with open("regmodel.pkl", "rb") as file:
    # Load the contents of the pickle file
    model = pickle.load(file)
```

```python
X = ds.copy(deep=True)
print(X.columns)
```

```python
# Run the prediction
y_pred= model.predict(X) 
```

```python
# Add a new column "Predicted" to the original DataFrame X
X['Predicted'] = y_pred
X
```
